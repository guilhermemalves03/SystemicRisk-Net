```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(7, 4.5), dpi=300)


cohorts = ['STEM (n=3)', 'Business (n=2)', 'Design (n=1)']
task_accuracy = [4.6, 4.5, 3.0]    
aesthetic_score = [3.8, 4.0, 4.0]  
typography_score = [4.2, 4.0, 2.0] 

x = np.arange(len(cohorts))
width = 0.25


rects1 = ax.bar(x - width, task_accuracy, width, label='Functional Tasks (1-5)', color='#3498db')
rects2 = ax.bar(x, aesthetic_score, width, label='General Aesthetics (1-5)', color='#2ecc71')
rects3 = ax.bar(x + width, typography_score, width, label='Typography Clarity (1-5)', color='#e74c3c')


ax.set_ylabel('Evaluation Score (1 - 5)', fontsize=11, fontname='sans-serif')
ax.set_title('SystemicRisk-Net Dynamic Evaluation Matrix', fontsize=13, pad=15, fontweight='bold', fontname='serif')
ax.set_xticks(x)
ax.set_xticklabels(cohorts, fontsize=10)
ax.set_ylim(0, 5.8)


ax.grid(axis='y', linestyle='--', alpha=0.15)
ax.set_axisbelow(True)


for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)


ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=3, frameon=False, fontsize=9)


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9, color='#ffffff', fontweight='bold')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.tight_layout()

```
